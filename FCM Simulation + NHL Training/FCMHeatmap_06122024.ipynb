```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_weight_matrix(matrix, matrix_name="Weight Matrix"):
    """
    Analyze and visualize a weight matrix.
    """
    # Statistics
    non_zero_weights = matrix[np.abs(matrix) > 1e-5]  # Exclude near-zero weights
    sparsity = 1 - len(non_zero_weights) / matrix.size
    mean_weight = non_zero_weights.mean()
    std_weight = non_zero_weights.std()
    min_weight = non_zero_weights.min()
    max_weight = non_zero_weights.max()

    # Print statistics
    print(f"Analysis for {matrix_name}:")
    print(f"Sparsity: {sparsity:.2%}")
    print(f"Mean Weight: {mean_weight:.4f}")
    print(f"Standard Deviation: {std_weight:.4f}")
    print(f"Minimum Weight: {min_weight:.4f}")
    print(f"Maximum Weight: {max_weight:.4f}")

    # Heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix, annot=False, cmap="coolwarm", cbar=True)
    plt.title(f"{matrix_name} Heatmap")
    plt.xlabel("Concepts")
    plt.ylabel("Concepts")
    plt.show()

    return {
        "Sparsity": sparsity,
        "Mean Weight": mean_weight,
        "Standard Deviation": std_weight,
        "Minimum Weight": min_weight,
        "Maximum Weight": max_weight,
    }

def compare_matrices(matrix1, matrix2, matrix1_name="Matrix 1", matrix2_name="Matrix 2"):
    """
    Compare two weight matrices and visualize their differences.
    """
    # Element-wise difference
    diff_matrix = matrix1 - matrix2

    # Statistics
    mean_diff = np.mean(diff_matrix)
    std_diff = np.std(diff_matrix)
    max_diff = np.max(diff_matrix)
    min_diff = np.min(diff_matrix)

    print(f"Comparison between {matrix1_name} and {matrix2_name}:")
    print(f"Mean Difference: {mean_diff:.4f}")
    print(f"Standard Deviation of Difference: {std_diff:.4f}")
    print(f"Maximum Difference: {max_diff:.4f}")
    print(f"Minimum Difference: {min_diff:.4f}")

    # Heatmap for differences
    plt.figure(figsize=(10, 8))
    sns.heatmap(diff_matrix, annot=False, cmap="coolwarm", cbar=True)
    plt.title(f"Difference Between {matrix1_name} and {matrix2_name}")
    plt.xlabel("Concepts")
    plt.ylabel("Concepts")
    plt.show()

    return {
        "Mean Difference": mean_diff,
        "Standard Deviation of Difference": std_diff,
        "Maximum Difference": max_diff,
        "Minimum Difference": min_diff,
    }

# Example usage:
# Load your matrices as numpy arrays (replace with your file paths)
data_matrix = pd.read_csv("correlation_weight_matrix_30112024_cleared.csv", index_col=0).values
expert_matrix = pd.read_csv("Mentalmodeler_extract_30112024.csv", index_col=0).values

# Analyze each matrix
data_stats = analyze_weight_matrix(data_matrix, matrix_name="Data-Driven Matrix")
expert_stats = analyze_weight_matrix(expert_matrix, matrix_name="Expert-Assigned Matrix")

# Compare matrices
comparison_stats = compare_matrices(data_matrix, expert_matrix, 
                                     matrix1_name="Data-Driven Matrix", 
                                     matrix2_name="Expert-Assigned Matrix")
```
